### 평균기온 그래프

```python
f = open('기상관측_서울_20181004141633.csv', encoding='cp949')

f.readline()

t = []
m = []

for line in f:
    l = line.strip().split(',')
    t.append(float(l[2]))
    m.append(float(l[24]))

f.close()
```

```python
plt.plot(t)
plt.plot(m)
```
